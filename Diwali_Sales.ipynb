```python
d1=sns.countplot(x='Gender',data=df)
for i in d1.containers:
  d1.bar_label(i) # shows the value also
```

```python
d2=sns.countplot(data=df,x='Age Group')
for i in d2.containers:
  d2.bar_label(i)
```

```python
d2=sns.countplot(x='Age Group',data=df,hue='Gender')
for i in d2.containers:
  d2.bar_label(i)
```

```python
#group by using age vs sales with bar graph
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age)
```

Total Orders/sales from top 5 states

```python

sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(5)

sns.set(rc={'figure.figsize':(9,3)})
sns.barplot(data = sales_state, x = 'State',y= 'Orders')
```

Total Amount/sales from top 5 states

```python
# total amount/sales from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(5)

sns.set(rc={'figure.figsize':(9,3)})
sns.barplot(data = sales_state, x = 'State',y= 'Amount')
```
